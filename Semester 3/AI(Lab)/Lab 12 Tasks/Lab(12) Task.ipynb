# Data Preprocessing

## At first, I'll fill missing values

```python
encoded_df = df.copy()
```

```python
encoded_df1 = df1.copy()
```

```python
encoded_df2 = df2.copy()
```

## Filling object values

```python
def fillobj(data,col):
    for i in col:
        data[i] = data[i].fillna(data[i].mode()[0])
```

```python
col1 = ['Cabin','Embarked']
fillobj(encoded_df1,col1)
col2 = ['Cabin']
fillobj(encoded_df2,col2)
```

```python
col1 = ['Age']
fillobj(encoded_df1,col1)
col2 = ['Age',"Fare"]
fillobj(encoded_df2,col2)
```

```python
encoded_df1.isnull().sum()
```

```python
encoded_df2.isnull().sum()
```

### Hence, Missing values are filled.

## Dropping useless column

```python
def dropcol(data,col):
    for i in col:
        data.drop(i,axis = 1,inplace = True)
```

```python
col = ['PassengerId','Name','Ticket']
dropcol(encoded_df1,col)
dropcol(encoded_df2,col)
```

```python
col2 = ['PassengerId']
dropcol(encoded_df,col2)
```

### Useless columns dropped

## Converting columns to same data type

```python
def to_int64(data,col):
    for i in col:
        data[i] = data[i].astype('int64')
```

```python
col = ['Age','Fare']
to_int64(encoded_df1,col)
to_int64(encoded_df2,col)
```

## Encoding multiple categorical columns

```python
def encode_col_with_ordinal(data, cols):
    encoder = OrdinalEncoder()
    data[cols] = encoder.fit_transform(data[cols])
    return data
```

```python
cols = ['Sex', 'Cabin', 'Embarked']
encoded_train_df = encode_col_with_ordinal(encoded_df1, cols)
encoded_test_X = encode_col_with_ordinal(encoded_df2, cols)
```
